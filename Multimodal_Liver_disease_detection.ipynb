```python
# Import Neccessary libraries
import os
import zipfile
import pandas as pd
import numpy as np
import cv2
from tqdm import tqdm
import albumentations as A
import pydicom
from io import BytesIO
from multiprocessing import Pool, cpu_count
import matplotlib.pyplot as plt
from collections import defaultdict

ZIP_PATH = "tcga_lihc.zip"
METADATA_PATH = "metadata.csv"
OUTPUT_DIR = "processed_images"

os.makedirs(OUTPUT_DIR, exist_ok=True)
df = pd.read_csv(METADATA_PATH)

# Preprocessing
transform = A.Compose([
    A.CLAHE(clip_limit=2.0, p=0.7),
    A.RandomBrightnessContrast(p=0.3),
    A.RandomGamma(p=0.3),
    A.GaussNoise(var_limit=(5.0, 20.0), p=0.2),
    A.HorizontalFlip(p=0.5),
    A.ShiftScaleRotate(
        shift_limit=0.03,
        scale_limit=0.05,
        rotate_limit=5,
        p=0.4
    ),
    A.Resize(224, 224),
    A.Normalize(mean=(0.5,), std=(0.5,))
])

# Helper functions
def dicom_to_hu(ds):
    image = ds.pixel_array.astype(np.int16)
    intercept = getattr(ds, "RescaleIntercept", 0)
    slope = getattr(ds, "RescaleSlope", 1)
    return image * slope + intercept

def window_image(img, level=50, width=150):
    min_val = level - width // 2
    max_val = level + width // 2
    img = np.clip(img, min_val, max_val)
    img = (img - min_val) / (max_val - min_val)
    return (img * 255).astype(np.uint8)
    
#Advanced preprocessing
def normalize_hu(image):
    image = np.clip(image, -1000, 400)
    image = (image + 1000) / 1400
    return image    
def extract_roi(image):
    img_uint8 = (image * 255).astype(np.uint8)
    _, mask = cv2.threshold(
        img_uint8, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )
    kernel = np.ones((5, 5), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    roi = cv2.bitwise_and(img_uint8, img_uint8, mask=mask)
    return roi, mask
def denoise(image):
    return cv2.bilateralFilter(image, d=5, sigmaColor=75, sigmaSpace=75)
def compute_tbr_and_weight(image, mask):
    tissue = image[mask > 0]
    background = image[mask == 0]
    if len(background) == 0:
        tbr = 0
    else:
        tbr = np.mean(tissue) / (np.mean(background) + 1e-6)
    weighted = image * (1 + tbr)
    weighted = np.clip(weighted, 0, 255).astype(np.uint8)
    return weighted, tbr
    
print(" Indexing ZIP...")
zip_ref = zipfile.ZipFile(ZIP_PATH, 'r')
zip_files = zip_ref.namelist()
index = []
for f in zip_files:
    if f.endswith(".dcm"):
        parts = f.split("/")
        if len(parts) >= 5:
            patient = parts[1]
            study = parts[2]
            series = parts[3]
            index.append((patient, study, series, f))
print(f"Indexed {len(index)} DICOM files")
lookup = defaultdict(list)
for patient, study, series, path in index:
    lookup[(patient, study, series)].append(path)
print(f" Unique series: {len(lookup)}")

def process_series(args):
    patient, study, series, file_list = args
    local_zip = zipfile.ZipFile(ZIP_PATH, 'r')
    records = []
    save_folder = os.path.join(OUTPUT_DIR, patient, study, series)
    os.makedirs(save_folder, exist_ok=True)
    for file_path in file_list:
        try:
            with local_zip.open(file_path) as f:
                dicom_bytes = f.read()
            ds = pydicom.dcmread(BytesIO(dicom_bytes))
            image = dicom_to_hu(ds)
            image = normalize_hu(image)
            roi, mask = extract_roi(image)
            roi = denoise(roi)
            roi, tbr = compute_tbr_and_weight(roi, mask)
            augmented = transform(image=roi)
            processed = augmented["image"]
            filename = os.path.basename(file_path).replace(".dcm", ".png")
            save_path = os.path.join(save_folder, filename)
            cv2.imwrite(save_path, processed)
            records.append({
                "PatientID": patient,
                "StudyInstanceUID": study,
                "SeriesInstanceUID": series,
                "image_path": save_path,
                "TBR": tbr
            })
        except Exception as e:
            print(f"Error: {file_path} -> {e}")
    local_zip.close()
    return records

tasks = []

for _, row in df.iterrows():
    key = (
        str(row["PatientID"]),
        str(row["StudyInstanceUID"]),
        str(row["SeriesInstanceUID"])
    )

    if key in lookup:
        tasks.append((key[0], key[1], key[2], lookup[key]))

print(f" Total series to process: {len(tasks)}")
num_workers = max(1, cpu_count() - 1)
print(f" Using {num_workers} CPU cores")

all_records = []

with Pool(num_workers) as pool:
    for result in tqdm(pool.imap_unordered(process_series, tasks), total=len(tasks)):
        all_records.extend(result)

mapping_path = os.path.join(OUTPUT_DIR, "mapping.csv")
pd.DataFrame(all_records).to_csv(mapping_path, index=False)

print("\n DONE!")
print(f"Images: {OUTPUT_DIR}")
print(f" Mapping: {mapping_path}")

def visualize_pipeline(sample_path):
    with zip_ref.open(sample_path) as f:
        dicom_bytes = f.read()
    ds = pydicom.dcmread(BytesIO(dicom_bytes))
    original = dicom_to_hu(ds)
    windowed = window_image(original)
    norm = normalize_hu(original)
    roi, mask = extract_roi(norm)
    den = denoise(roi)
    weighted, tbr = compute_tbr_and_weight(den, mask)
    plt.figure(figsize=(12, 8))
    plt.subplot(2,3,1)
    plt.title("Original (Windowed)")
    plt.imshow(windowed, cmap='gray')
    plt.subplot(2,3,2)
    plt.title("Normalized")
    plt.imshow(norm, cmap='gray')
    plt.subplot(2,3,3)
    plt.title("ROI")
    plt.imshow(roi, cmap='gray')
    plt.subplot(2,3,4)
    plt.title("Mask")
    plt.imshow(mask, cmap='gray')
    plt.subplot(2,3,5)
    plt.title("Denoised")
    plt.imshow(den, cmap='gray')
    plt.subplot(2,3,6)
    plt.title(f"Final (TBR={tbr:.2f})")
    plt.imshow(weighted, cmap='gray')
    plt.tight_layout()
    plt.show()
visualize_pipeline(index[0][3])
```

```python
import sys
sys.path.insert(0, "/nfsshare/users/P127156020/my_packages")

import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.models as models
from tqdm import tqdm
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model = models.resnet50(pretrained=True).to(device)
model.eval()

feature_extractor = nn.Sequential(*list(model.children())[:-1]).to(device)
gradients = []
activations = []

def save_gradients(module, grad_input, grad_output):
    gradients.append(grad_output[0])

def save_activations(module, input, output):
    activations.append(output)

model.layer4.register_forward_hook(save_activations)
model.layer4.register_full_backward_hook(save_gradients)
def prepare_image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (224, 224))

    img = np.stack([img, img, img], axis=-1)
    img = img / 255.0
    img = (img - 0.5) / 0.5

    img = torch.tensor(img, dtype=torch.float32).permute(2,0,1)
    return img.unsqueeze(0).to(device)

def extract_features(img_tensor):
    with torch.no_grad():
        feat = feature_extractor(img_tensor)
    return feat.squeeze().detach().cpu().numpy()

#GRAD-CAM
def generate_gradcam(img_tensor):
    gradients.clear()
    activations.clear()
    output = model(img_tensor)
    class_idx = output.argmax()
    model.zero_grad()
    output[0, class_idx].backward()
    grad = gradients[0]
    act = activations[0]
    weights = grad.mean(dim=(2,3), keepdim=True)
    cam = (weights * act).sum(dim=1)
    cam = F.relu(cam)
    return cam.squeeze().detach().cpu().numpy()

def visualize_feature_maps(img_tensor):
    x = img_tensor
    for layer in list(model.children())[:5]:
        x = layer(x)

    fmap = x[0].detach().cpu()

    plt.figure(figsize=(8,8))
    for i in range(min(16, fmap.shape[0])):
        plt.subplot(4,4,i+1)
        plt.imshow(fmap[i], cmap='gray')
        plt.axis('off')
    plt.suptitle("Feature Maps")
    plt.show()

IMAGE_DIR = "processed_images"

all_images = []
for root, _, files in os.walk(IMAGE_DIR):
    for f in files:
        if f.endswith(".png"):
            all_images.append(os.path.join(root, f))

print("Total images:", len(all_images))

features = []

print(" Extracting features...")

for path in tqdm(all_images):
    try:
        img_tensor = prepare_image(path)
        feat = extract_features(img_tensor)
        features.append(feat)

        # Prevent GPU memory issues
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    except Exception as e:
        print("Error:", path, e)

features = np.array(features)
print("Feature shape:", features.shape)

#PCA
pca = PCA(n_components=2)
pca_out = pca.fit_transform(features)
plt.figure(figsize=(7,6))
plt.scatter(pca_out[:,0], pca_out[:,1], s=5)
plt.title("PCA - All Images")
plt.show()

#t-SNE
tsne = TSNE(n_components=2, perplexity=30)
tsne_out = tsne.fit_transform(features)

plt.figure(figsize=(7,6))
plt.scatter(tsne_out[:,0], tsne_out[:,1], s=5)
plt.title("t-SNE - All Images")
plt.show()

sample_paths = random.sample(all_images, min(5, len(all_images)))
for path in sample_paths:
    print("\nAnalyzing:", path)
    img_tensor = prepare_image(path)
    visualize_feature_maps(img_tensor)
    cam = generate_gradcam(img_tensor)
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    cam = cv2.resize(cam, (224,224))
    cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-6)
    plt.imshow(img, cmap='gray')
    plt.imshow(cam, cmap='jet', alpha=0.5)
    plt.title("Grad-CAM")
    plt.axis('off')
    plt.show()
```

```python
#Multi-modal
import sys
sys.path.insert(0, "/nfsshare/users/P127156020/my_packages")

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms

import pandas as pd
import numpy as np
from PIL import Image
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if device.type == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))

torch.backends.cudnn.benchmark = True

df_mapping = pd.read_csv("processed_images/mapping.csv")
df_clinical = pd.read_csv("clinical_filtered.csv")
df_mapping.columns = df_mapping.columns.str.strip()
df_clinical.columns = df_clinical.columns.str.strip()
df = pd.merge(
    df_mapping,
    df_clinical,
    left_on="PatientID",
    right_on="cases.submitter_id"
)
#preprocessing
df.drop(columns=["cases.submitter_id"], inplace=True, errors="ignore")

df = df.dropna(subset=["diagnoses.primary_diagnosis"])
df["target"] = pd.factorize(df["diagnoses.primary_diagnosis"])[0]

num_classes = df["target"].nunique()
print("Classes:", num_classes)

cols_to_drop = [
    "demographic.days_to_death",
    "diagnoses.days_to_last_follow_up",
    "diagnoses.progression_or_recurrence"
]

df.drop(columns=cols_to_drop, inplace=True, errors="ignore")
clinical_cols = [col for col in df.columns if col not in ["image_path", "target"]]
# Convert safely
for col in clinical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
# Fill NaNs
df[clinical_cols] = df[clinical_cols].fillna(df[clinical_cols].median())
df[clinical_cols] = df[clinical_cols].fillna(0)
# Convert type
df[clinical_cols] = df[clinical_cols].astype(np.float32)

scaler = StandardScaler()
df[clinical_cols] = scaler.fit_transform(df[clinical_cols])

print("Clinical features:", len(clinical_cols))



class MultiModalDataset(Dataset):
    def __init__(self, df):
        self.df = df.reset_index(drop=True)
        self.transform = transforms.Compose([
            transforms.Resize((64,64), antialias=True),
            transforms.ToTensor()
        ])
        self.clinical = torch.tensor(
            self.df[clinical_cols].values,
            dtype=torch.float32
        )

        self.labels = torch.tensor(
            self.df["target"].values,
            dtype=torch.long
        )
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        try:
            img = Image.open(row["image_path"]).convert("RGB")
        except:
            img = Image.fromarray(np.zeros((64,64,3), dtype=np.uint8))

        img = self.transform(img)

        return img, self.clinical[idx], self.labels[idx]


#splitting
train_df, val_df = train_test_split(
    df,
    test_size=0.1,
    stratify=df["target"],
    random_state=42
)

train_dataset = MultiModalDataset(train_df)
val_dataset   = MultiModalDataset(val_df)

print(" Dataset ready")

train_loader = DataLoader(
    train_dataset,
    batch_size=256,
    shuffle=True,
    num_workers=16,
    pin_memory=True,
    persistent_workers=True,
    prefetch_factor=4
)

val_loader = DataLoader(
    val_dataset,
    batch_size=256,
    num_workers=16,
    pin_memory=True,
    persistent_workers=True,
    prefetch_factor=4
)
class MultiModalNet(nn.Module):
    def __init__(self, num_clinical, num_classes):
        super().__init__()
        self.cnn = models.mobilenet_v3_small(weights="IMAGENET1K_V1")
        # Freeze CNN
        for param in self.cnn.features.parameters():
            param.requires_grad = False

        self.cnn.classifier = nn.Sequential(
            nn.Linear(576, 128),
            nn.ReLU()
        )
        self.clinical_net = nn.Sequential(
            nn.Linear(num_clinical, 32),
            nn.ReLU()
        )
        self.fc = nn.Sequential(
            nn.Linear(128 + 32, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, num_classes)
        )
    def forward(self, img, clin):
        img_feat = self.cnn(img)
        clin = torch.clamp(clin, -5, 5)
        clin_feat = self.clinical_net(clin)
        out = torch.cat((img_feat, clin_feat), dim=1)
        out = torch.clamp(out, -10, 10)
        return self.fc(out)
model = MultiModalNet(len(clinical_cols), num_classes).to(device)
model = torch.compile(model)

criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.Adam(model.parameters(), lr=5e-6)

scaler = torch.amp.GradScaler('cuda')
def train_epoch(loader):
    model.train()
    total_loss = 0

    for img, clin, y in loader:
        img = img.to(device, non_blocking=True)
        clin = clin.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        optimizer.zero_grad()

        with torch.amp.autocast('cuda'):
            out = model(img, clin)
            loss = criterion(out, y)

        if torch.isnan(loss):
            print(" NaN LOSS DETECTED")
            return np.nan

        scaler.scale(loss).backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    return total_loss / len(loader)


def evaluate(loader):
    model.eval()
    correct = 0

    with torch.no_grad():
        for img, clin, y in loader:
            img = img.to(device, non_blocking=True)
            clin = clin.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            pred = model(img, clin).argmax(1)
            correct += (pred == y).sum().item()

    return correct / len(loader.dataset)
for epoch in range(5):
    loss = train_epoch(train_loader)
    acc = evaluate(val_loader)

    print(f"Epoch {epoch+1}: Loss={loss:.4f}, Val Acc={acc:.4f}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

class_names = df['target'].unique()
num_classes = len(class_names)
model.eval()
all_probs = []
all_labels = []
with torch.no_grad():
    for img, clin, y in val_loader:
        img = img.to(device, non_blocking=True)
        clin = clin.to(device, non_blocking=True)
        with torch.cuda.amp.autocast():
            outputs = model(img, clin)
            probs = torch.softmax(outputs, dim=1)
        all_probs.append(probs.detach().cpu())
        all_labels.append(y.detach().cpu())
all_probs = torch.cat(all_probs).numpy()
all_labels = torch.cat(all_labels).numpy()
print("Detected classes:", num_classes)
print("Class names:", class_names)
y_bin = label_binarize(all_labels, classes=np.arange(num_classes))
fpr, tpr, roc_auc = {}, {}, {}
for i, class_name in enumerate(class_names):
    fpr[i], tpr[i], _ = roc_curve(y_bin[:, i], all_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
fpr["micro"], tpr["micro"], _ = roc_curve(
    y_bin.ravel(), all_probs.ravel()
)
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
all_auc = [roc_auc[i] for i in range(num_classes)]
roc_auc["macro"] = np.mean(all_auc)
print("\nAUC Scores:")
for i, class_name in enumerate(class_names):
    print(f"{class_name}: {roc_auc[i]:.4f}")
print(f"Micro Avg: {roc_auc['micro']:.4f}")
print(f"Macro Avg: {roc_auc['macro']:.4f}")


plt.figure(figsize=(8, 6))
plt.plot(
    fpr["micro"], tpr["micro"],
    label=f"Micro-average (AUC = {roc_auc['micro']:.2f})",
    linewidth=3
)
for i, class_name in enumerate(class_names):
    plt.plot(
        fpr[i], tpr[i],
        label=f"{class_name} (AUC = {roc_auc[i]:.2f})"
    )

plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Multi-class)")
plt.legend()
plt.grid()

plt.show()
```

```python
import os
import numpy as np
import pandas as pd

from hmmlearn import hmm
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.utils.class_weight import compute_class_weight
from PIL import Image

print("CPU-only pipeline (stable)")
df_mapping = pd.read_csv("processed_images/mapping.csv")
df_clinical = pd.read_csv("clinical_filtered.csv")

df_mapping.columns = df_mapping.columns.str.strip()
df_clinical.columns = df_clinical.columns.str.strip()

df = pd.merge(
    df_mapping,
    df_clinical,
    left_on="PatientID",
    right_on="cases.submitter_id",
    how="inner"
)

df.drop(columns=["cases.submitter_id"], inplace=True, errors="ignore")

labels, class_names = pd.factorize(df["diagnoses.primary_diagnosis"])
df["target"] = labels

num_classes = 3
print("Classes:", num_classes)

drop_cols = [
    "demographic.days_to_death",
    "diagnoses.days_to_last_follow_up",
    "diagnoses.progression_or_recurrence"
]

df.drop(columns=drop_cols, inplace=True, errors="ignore")
clinical_cols = [c for c in df.columns if c not in ["image_path", "target", "PatientID"]]
for c in clinical_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df[clinical_cols] = df[clinical_cols].fillna(0).astype(np.float32)

train_df, val_df = train_test_split(
    df,
    test_size=0.1,
    stratify=df["target"],  
    random_state=42
)

print("Train distribution:\n", train_df["target"].value_counts())
print("\nVal distribution:\n", val_df["target"].value_counts())


CACHE_DIR = "cache_seq"
os.makedirs(CACHE_DIR, exist_ok=True)

def image_to_sequence_cached(path):
    fname = os.path.join(CACHE_DIR, os.path.basename(path) + ".npy")

    if os.path.exists(fname):
        return np.load(fname)

    try:
        img = Image.open(path).convert("RGB").resize((128,128))
        img = np.array(img) / 255.0
    except:
        img = np.zeros((128,128,3))

    patches = []
    for i in range(0, 128, 64):
        for j in range(0, 128, 64):
            patch = img[i:i+64, j:j+64]
            patches.append(patch.mean(axis=(0,1)))

    seq = np.array(patches, dtype=np.float32)
    np.save(fname, seq)

    return seq


def train_hmm(df):
    models = {}
    for c in range(num_classes):
        paths = df[df["target"] == c]["image_path"].values
        sequences, lengths = [], []
        for p in paths[:100]:  # speed limit
            seq = image_to_sequence_cached(p)
            sequences.append(seq)
            lengths.append(len(seq))
        X = np.concatenate(sequences)
        model = hmm.GaussianHMM(
            n_components=1,
            covariance_type="diag",
            n_iter=5)
        model.fit(X, lengths)
        models[c] = model
    return models
print("Training HMM...")
hmm_models = train_hmm(train_df)
def compute_hmm_features(df, save_path):
    if os.path.exists(save_path):
        return np.load(save_path)
    feats = []
    for i, path in enumerate(df["image_path"]):
        seq = image_to_sequence_cached(path)
        row = [
            m.score(seq) if len(seq) else 0.0
            for m in hmm_models.values()
        ]
        feats.append(row)
        if i % 2000 == 0:
            print(f"{i}/{len(df)}")
    feats = np.array(feats, dtype=np.float32)
    np.save(save_path, feats)
    return feats
train_hmm_feat = compute_hmm_features(train_df, "train_hmm.npy")
val_hmm_feat   = compute_hmm_features(val_df, "val_hmm.npy")

X_train = np.concatenate([
    train_hmm_feat,
    train_df[clinical_cols].values
], axis=1)

X_val = np.concatenate([
    val_hmm_feat,
    val_df[clinical_cols].values
], axis=1)

y_train = train_df["target"].values
y_val   = val_df["target"].values

classes = np.unique(y_train)
weights = compute_class_weight("balanced", classes=classes, y=y_train)

class_weights = dict(zip(classes, weights))
print("Class Weights:", class_weights)

clf = MLPClassifier(
    hidden_layer_sizes=(128, 64),   # deeper network
    activation='relu',
    solver='adam',
    batch_size=256,
    max_iter=30,
    alpha=1e-4,       
    early_stopping=True, 
    validation_fraction=0.1,
    n_iter_no_change=5,
    verbose=True
)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_val)

print("\nAccuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report:\n", classification_report(y_val, y_pred))
```

```python
#Leakage test
import hashlib
from sklearn.linear_model import LogisticRegression

print("\n Detecting clinical leakage...")
leakage_cols = []
for col in clinical_cols:
    if train_df[col].nunique() < 2:
        continue
    # Correlation check
    corr = abs(np.corrcoef(train_df[col], train_df["target"])[0,1])
    if not np.isnan(corr) and corr > 0.8:
        leakage_cols.append(col)
        continue
    # Single-feature prediction test (fast)
    try:
        model = LogisticRegression(max_iter=100)
        model.fit(train_df[[col]], train_df["target"])
        acc = model.score(val_df[[col]], val_df["target"])
        if acc > 0.8:
            leakage_cols.append(col)
    except:
        pass

leakage_cols = list(set(leakage_cols))
print("\n Clinical Leakage Columns Detected:")
print(leakage_cols[:20])
train_df.drop(columns=leakage_cols, inplace=True, errors="ignore")
val_df.drop(columns=leakage_cols, inplace=True, errors="ignore")
# Update clinical cols
clinical_cols = [c for c in train_df.columns if c not in ["image_path","target"]]
print("\n Remaining clinical features:", len(clinical_cols))
print("\n Checking patient overlap...")
overlap = set(train_df.PatientID) & set(val_df.PatientID)
print("Patient overlap:", len(overlap))
if len(overlap) > 0:
    print(" Removing overlapping patients...")
    val_df = val_df[~val_df.PatientID.isin(overlap)].copy()
print(" Patient leakage handled")


print("\n Checking exact duplicate images...")
def hash_img(p):
    try:
        return hashlib.md5(open(p,'rb').read()).hexdigest()
    except:
        return None
train_df["img_hash"] = train_df["image_path"].apply(hash_img)
val_df["img_hash"]   = val_df["image_path"].apply(hash_img)
dup_hash = set(train_df.img_hash) & set(val_df.img_hash)
print("Exact duplicates:", len(dup_hash))
if len(dup_hash) > 0:
    print("Removing duplicate images from validation...")
    val_df = val_df[~val_df.img_hash.isin(dup_hash)].copy()
print("Exact image leakage handled")


print("\n Checking perceptual duplicates...")
try:
    import imagehash
    from PIL import Image
    train_hashes = {}
    val_hashes = {}
    # Sample for speed
    for p in train_df.image_path.sample(min(200, len(train_df))):
        try:
            train_hashes[p] = imagehash.phash(Image.open(p))
        except:
            pass
    for p in val_df.image_path.sample(min(200, len(val_df))):
        try:
            val_hashes[p] = imagehash.phash(Image.open(p))
        except:
            pass
    similar = []
    for vp, vh in val_hashes.items():
        for th in train_hashes.values():
            if abs(vh - th) < 5:  # similarity threshold
                similar.append(vp)
                break

    print("Perceptual duplicates:", len(similar))
    if len(similar) > 0:
        print(" Removing perceptual duplicates from validation...")
        val_df = val_df[~val_df.image_path.isin(similar)].copy()
    print(" Perceptual leakage handled")
except:
    print(" imagehash not installed, skipping perceptual check")


train_df.drop(columns=["img_hash"], inplace=True, errors="ignore")
val_df.drop(columns=["img_hash"], inplace=True, errors="ignore")
print("\n======================")
print(" FINAL DATA STATUS")
print("======================")
print("Train samples:", len(train_df))
print("Val samples:", len(val_df))
print("Clinical features:", len(clinical_cols))
print("Removed leakage columns:", len(leakage_cols))
```

```python

import sys, os
sys.path.insert(0, "/nfsshare/users/P127156020/my_packages")

os.environ["CUDA_VISIBLE_DEVICES"] = "1"

import torch, torch.nn as nn, torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms

import pandas as pd, numpy as np
from PIL import Image

# Implementation of Multi-modal after Leakage test
from sklearn.model_selection import GroupShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.base import BaseEstimator, TransformerMixin
from statsmodels.stats.outliers_influence import variance_inflation_factor
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
df_mapping = pd.read_csv("processed_images/mapping.csv")
df_clinical = pd.read_csv("clinical_filtered.csv")

df = pd.merge(df_mapping, df_clinical,
              left_on="PatientID",
              right_on="cases.submitter_id")
df.drop(columns=["cases.submitter_id"], inplace=True, errors="ignore")
df = df.dropna(subset=["diagnoses.primary_diagnosis"])
df["target"] = pd.factorize(df["diagnoses.primary_diagnosis"])[0]
hash_df = pd.read_csv("image_hash_cache.csv")
df = df.drop_duplicates(subset=["image_path"])
df = df.merge(hash_df, on="image_path", how="left")
df = df.dropna(subset=["img_hash"])
df = df.drop_duplicates(subset=["img_hash"])

#split
gss = GroupShuffleSplit(test_size=0.1, n_splits=1, random_state=42)
train_idx, val_idx = next(gss.split(df, groups=df["PatientID"]))

train_df = df.iloc[train_idx].copy()
val_df   = df.iloc[val_idx].copy()
exclude = ["image_path","PatientID","target",
           "diagnoses.primary_diagnosis","img_hash"]
clinical_cols = [c for c in df.columns if c not in exclude]
# convert to numeric
for col in clinical_cols:
    train_df[col] = pd.to_numeric(train_df[col], errors="coerce")
    val_df[col]   = pd.to_numeric(val_df[col], errors="coerce")
X_train = train_df[clinical_cols]
y_train = train_df["target"]
X_val = val_df[clinical_cols]
y_val = val_df["target"]


class VIFSelector(BaseEstimator, TransformerMixin):
    def __init__(self, thresh=5.0):
        self.thresh = thresh

    def fit(self, X, y=None):
        X = pd.DataFrame(X)
        cols = list(range(X.shape[1]))
        while len(cols) > 1:
            vif = [variance_inflation_factor(X.iloc[:, cols].values, i)
                   for i in range(len(cols))]
            if max(vif) > self.thresh:
                cols.pop(vif.index(max(vif)))
            else:
                break
        self.selected_ = cols
        return self
    def transform(self, X):
        return np.array(X)[:, self.selected_]

#MI and VIF
pipe = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler()),
    ("mi", SelectKBest(mutual_info_classif, k=20)),
    ("vif", VIFSelector(thresh=5.0))
])
pipe.fit(X_train, y_train)
X_train_trans = pipe.transform(X_train)
X_val_trans   = pipe.transform(X_val)
class MultiModalDataset(Dataset):
    def __init__(self, df, clin_data):
        self.df = df.reset_index(drop=True)
        self.clin = torch.tensor(clin_data, dtype=torch.float32)
        self.labels = torch.tensor(df["target"].values, dtype=torch.long)

        self.transform = transforms.Compose([
            transforms.Resize((48,48)),
            transforms.ToTensor()
        ])
        self.images = []
        for path in df["image_path"]:
            try:
                img = Image.open(path).convert("RGB")
            except:
                img = Image.fromarray(np.zeros((48,48,3), dtype=np.uint8))
            self.images.append(self.transform(img))

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        return self.images[idx], self.clin[idx], self.labels[idx]
train_loader = DataLoader(
    MultiModalDataset(train_df, X_train_trans),
    batch_size=32,
    shuffle=True,
    num_workers=4,
    pin_memory=True
)
val_loader = DataLoader(
    MultiModalDataset(val_df, X_val_trans),
    batch_size=32,
    num_workers=4
)

class Net(nn.Module):
    def __init__(self, clin_dim):
        super().__init__()

        self.cnn = models.mobilenet_v3_small(weights="IMAGENET1K_V1")

        for p in self.cnn.parameters():
            p.requires_grad = False

        self.cnn.classifier = nn.Sequential(
            nn.Linear(576, 32),
            nn.ReLU()
        )
        self.clin = nn.Sequential(
            nn.Linear(clin_dim, 8),
            nn.ReLU()
        )
        self.fc = nn.Sequential(
            nn.Linear(32+8, 32),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(32, len(np.unique(y_train)))
        )

    def forward(self, img, clin):
        x1 = self.cnn(img)
        x2 = self.clin(clin)
        return self.fc(torch.cat([x1, x2], dim=1))

model = Net(X_train_trans.shape[1]).to(device)

#optimizer
opt = optim.Adam(model.parameters(), lr=1e-4)
crit = nn.CrossEntropyLoss()

def evaluate(loader):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for img,clin,y in loader:
            img,clin,y = img.to(device), clin.to(device), y.to(device)
            pred = model(img,clin).argmax(1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return correct / total
best_acc = 0
for epoch in range(5):
    model.train()
    for img,clin,y in train_loader:
        img,clin,y = img.to(device), clin.to(device), y.to(device)

        opt.zero_grad()
        loss = crit(model(img,clin), y)
        loss.backward()
        opt.step()
    val_acc = evaluate(val_loader)
    print(f"Epoch {epoch+1} | Val Acc: {val_acc:.4f}")

    if val_acc > best_acc:
        best_acc = val_acc
        torch.save(model.state_dict(), "best_pipeline_model.pth")
print(" Best Accuracy:", best_acc)
```

```python
import os
import numpy as np
import pandas as pd

from hmmlearn import hmm
from sklearn.model_selection import StratifiedKFold
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from PIL import Image

print(" Final Fast + Safe + No Leakage Pipeline")

df_mapping = pd.read_csv("processed_images/mapping.csv")
df_clinical = pd.read_csv("clinical_filtered.csv")

df = pd.merge(
    df_mapping,
    df_clinical,
    left_on="PatientID",
    right_on="cases.submitter_id",
    how="inner"
)

df.drop(columns=["cases.submitter_id"], inplace=True, errors="ignore")

labels, class_names = pd.factorize(df["diagnoses.primary_diagnosis"])
df["target"] = labels
num_classes = len(class_names)

print("Classes:", list(class_names))

exclude = ["image_path", "target", "PatientID", "diagnoses.primary_diagnosis"]
clinical_cols = [c for c in df.columns if c not in exclude]

for c in clinical_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df[clinical_cols] = df[clinical_cols].fillna(0).astype(np.float32)
BASE_CACHE = "cache_seq_base"
os.makedirs(BASE_CACHE, exist_ok=True)

BATCH_SIZE = 64
def load_batch_with_cache(paths):
    seq_batch = [None] * len(paths)
    to_compute = []
    compute_idx = []

    
    for i, p in enumerate(paths):
        fname = os.path.join(BASE_CACHE, os.path.basename(p) + ".npy")

        if os.path.exists(fname):
            try:
                seq_batch[i] = np.load(fname)
                continue
            except:
                os.remove(fname)

        to_compute.append(p)
        compute_idx.append(i)

    if len(to_compute) > 0:
        imgs = []

        for p in to_compute:
            try:
                img = Image.open(p).convert("RGB").resize((64,64))
                img = np.array(img, dtype=np.float32) / 255.0
            except:
                img = np.zeros((64,64,3), dtype=np.float32)

            imgs.append(img)

        imgs = np.stack(imgs)

        patches = []
        for i in [0, 32]:
            for j in [0, 32]:
                patch = imgs[:, i:i+32, j:j+32, :]
                patches.append(patch.mean(axis=(1,2)))

        seqs = np.stack(patches, axis=1)

        for idx, seq, path in zip(compute_idx, seqs, to_compute):
            fname = os.path.join(BASE_CACHE, os.path.basename(path) + ".npy")

            tmp = fname + ".tmp.npy"  
            np.save(tmp, seq)
            os.replace(tmp, fname)

            seq_batch[idx] = seq

    return np.array(seq_batch, dtype=np.float32)

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

fold_results = []
for fold, (train_idx, val_idx) in enumerate(skf.split(df, df["target"])):
    print(f"\n========== FOLD {fold+1} ==========")
    hmm_models = {}
    train_df = df.iloc[train_idx].copy()
    val_df   = df.iloc[val_idx].copy()
    print("Train:\n", train_df["target"].value_counts())
    print("Val:\n", val_df["target"].value_counts())
    train_df = train_df.sample(frac=0.3, random_state=42)
    val_df   = val_df.sample(frac=0.3, random_state=42)

    for c in range(num_classes):
        paths = train_df[train_df["target"] == c]["image_path"].values[:50]
        sequences, lengths = [], []
        for i in range(0, len(paths), BATCH_SIZE):
            batch = paths[i:i+BATCH_SIZE]
            seq_batch = load_batch_with_cache(batch)

            for seq in seq_batch:
                sequences.append(seq)
                lengths.append(len(seq))
        if len(sequences) == 0:
            continue
        X = np.concatenate(sequences)
        model = hmm.GaussianHMM(
            n_components=1,
            covariance_type="diag",
            n_iter=3
        )
        model.fit(X, lengths)
        hmm_models[c] = model

    def get_features_fast(df):
        feats = []
        paths = df["image_path"].values

        for i in range(0, len(paths), BATCH_SIZE):
            batch = paths[i:i+BATCH_SIZE]
            seq_batch = load_batch_with_cache(batch)

            for seq in seq_batch:
                row = [
                    hmm_models[c].score(seq) if c in hmm_models else 0.0
                    for c in range(num_classes)
                ]
                feats.append(row)

        return np.array(feats, dtype=np.float32)

    X_train = np.concatenate([
        get_features_fast(train_df),
        train_df[clinical_cols].values
    ], axis=1)

    X_val = np.concatenate([
        get_features_fast(val_df),
        val_df[clinical_cols].values
    ], axis=1)

    y_train = train_df["target"].values
    y_val   = val_df["target"].values

    clf = MLPClassifier(
        hidden_layer_sizes=(64,),
        max_iter=20,
        early_stopping=True,
        n_iter_no_change=3,
        batch_size=512,
        verbose=False
    )

    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_val)

    acc = accuracy_score(y_val, y_pred)
    print(f"Fold {fold+1} Accuracy:", acc)

    fold_results.append(acc)

print("\n Cross-Validation Accuracy:", np.mean(fold_results))
```
